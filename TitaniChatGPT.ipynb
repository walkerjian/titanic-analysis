```python
# Step 1: Handling Missing Values

# Imputing missing 'Age' values with the median
data['Age'].fillna(data['Age'].median(), inplace=True)

# Dropping the 'Cabin' column due to many missing values
data.drop(columns=['Cabin'], inplace=True)

# Filling missing 'Embarked' values with the most frequent value
most_frequent_embarked = data['Embarked'].mode()[0]
data['Embarked'].fillna(most_frequent_embarked, inplace=True)

# Checking if there are any more missing values
missing_values_after_imputation = data.isnull().sum()
missing_values_after_imputation
```

```python
# Step 2: Feature Engineering

# Creating a 'FamilySize' feature by combining 'SibSp' and 'Parch'
data['FamilySize'] = data['SibSp'] + data['Parch']

# Extracting titles from the 'Name' column
data['Title'] = data['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())

# Mapping common titles to broader categories
title_mapping = {
    'Mr': 'Mr',
    'Miss': 'Miss',
    'Mrs': 'Mrs',
    'Master': 'Master',
    'Dr': 'Officer',
    'Rev': 'Officer',
    'Mlle': 'Miss',
    'Major': 'Officer',
    'Col': 'Officer',
    'Ms': 'Mrs',
    'Mme': 'Mrs',
    'Lady': 'Royalty',
    'Sir': 'Royalty',
    'Capt': 'Officer',
    'Countess': 'Royalty',
    'Jonkheer': 'Royalty',
    'Don': 'Royalty',
    'Dona': 'Royalty'
}
data['Title'] = data['Title'].map(title_mapping)

# Previewing the first few rows to check the new features
data[['FamilySize', 'Title']].head()
```

```python
# Step 3: Encoding Categorical Variables

# Encoding 'Sex' column: 0 for male and 1 for female
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})

# One-hot encoding 'Embarked' and 'Title' columns
data = pd.get_dummies(data, columns=['Embarked', 'Title'], drop_first=True)

# Previewing the dataset after encoding
data.head()
```

```python
# Step 4: Scaling Numerical Features

from sklearn.preprocessing import StandardScaler

# Scaling 'Age' and 'Fare' using Standard Scaling
scaler = StandardScaler()
data[['Age', 'Fare']] = scaler.fit_transform(data[['Age', 'Fare']])

# Previewing the dataset after scaling
data[['Age', 'Fare']].head()
```

```python
# Step 5: Dropping Unnecessary Columns

# Dropping 'PassengerId', 'Name', and 'Ticket' columns
data_preprocessed = data.drop(columns=['PassengerId', 'Name', 'Ticket'])

# Previewing the final preprocessed dataset
data_preprocessed.head()
```

### perform a PCA within each class and show results graphically

Principal Component Analysis (PCA) is a dimensionality reduction technique used to transform high-dimensional data into a lower-dimensional form. It captures the most important information in the data by finding the principal components that explain the maximum variance.

Since we want to perform PCA within each class (Pclass: 1, 2, and 3), we'll apply PCA separately to the passengers in each class and visualize the results.

Let's start by performing PCA for each class and plotting the first two principal components:

Here are the results of Principal Component Analysis (PCA) performed separately for each passenger class:

- **Class 1 (Left)**: The plot shows the first two principal components for first-class passengers. There appears to be some separation between the survived and not survived groups, but it's not entirely distinct.
- **Class 2 (Middle)**: Similar to Class 1, there is some degree of separation between the two groups, but it is not clear-cut.
- **Class 3 (Right)**: The plot for third-class passengers shows more overlap between the survived and not survived groups, indicating higher complexity in this class.

This aligns with the complexity of the survival prediction task, where multiple features and their interactions play a role.

If you have any specific questions or need further analysis, please let me know!

```python
from sklearn.decomposition import PCA

# Function to perform PCA and plot the results for a given class
def plot_pca_for_class(pclass, ax):
    # Filtering data for the given class
    class_data = data_preprocessed[data_preprocessed['Pclass'] == pclass].drop(columns=['Pclass'])
    
    # Separating features and target variable
    X_class = class_data.drop(columns=['Survived'])
    y_class = class_data['Survived']
    
    # Applying PCA and reducing to two components
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(X_class)
    
    # Creating a DataFrame with the principal components
    pca_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
    pca_df['Survived'] = y_class.values
    
    # Plotting the two principal components
    sns.scatterplot(x='Principal Component 1', y='Principal Component 2', hue='Survived', data=pca_df, ax=ax)
    ax.set_title(f'PCA for Class {pclass}')
    ax.legend(title='Survived', labels=['Not Survived', 'Survived'])

# Plotting PCA for each class
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
for pclass, ax in zip([1, 2, 3], axes):
    plot_pca_for_class(pclass, ax)

plt.suptitle('Principal Component Analysis (PCA) Within Each Class')
plt.show()
```

This analysis provides insights into how certain features and their interactions create patterns in the reduced-dimensional space.

If you have any further questions or need additional insights, please let me know!

```python
# Function to perform PCA and return the loadings for a given class
def get_pca_loadings_for_class(pclass):
    # Filtering data for the given class
    class_data = data_preprocessed[data_preprocessed['Pclass'] == pclass].drop(columns=['Pclass'])
    
    # Separating features and target variable
    X_class = class_data.drop(columns=['Survived'])
    
    # Applying PCA and reducing to two components
    pca = PCA(n_components=2)
    pca.fit(X_class)
    
    # Creating a DataFrame with the loadings
    loadings_df = pd.DataFrame(pca.components_.T, columns=['Principal Component 1', 'Principal Component 2'], index=X_class.columns)
    
    return loadings_df

# Getting the PCA loadings for Class 2
loadings_class_2 = get_pca_loadings_for_class(2)

# Displaying the loadings for Class 2
loadings_class_2.sort_values(by='Principal Component 1', ascending=False)
```
